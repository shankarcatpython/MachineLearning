```
# prompt: read my_novel.pdf and convert them in to list of chunks
! pip install PyCryptodome
! pip install  pypdf

import  pypdf

from pypdf import PdfReader



# Open the PDF file in binary mode
with open('/content/my_novel.pdf', 'rb') as f:
  # Create a PDF reader object
  pdf_reader =  pypdf.PdfReader(f)

  # Initialize an empty list to store the chunks
  chunks = []

  print(pdf_reader.pages)

  # Loop over each page in the PDF document
  for page_num in range(len(pdf_reader.pages)):
    # Get the page object
    page = pdf_reader.pages[page_num]

    # Extract the text from the page
    text = page.extract_text()

    # Split the text into chunks of 1000 characters each
    for i in range(0, len(text), 1000):
      chunks.append(text[i:i+1000])

  # Print the list of chunks
  for chunk in chunks:
    print(len(chunk))
```

```
# prompt: convert chunks to embedding vector in python using spacy

import spacy

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Create an empty list to store the embedding vectors
embedding_vectors = []

# Loop over each chunk of text
for chunk in chunks:
  # Create a spaCy document object
  doc = nlp(chunk)

  # Get the embedding vector of the document
  embedding_vector = doc.vector

  # Append the embedding vector to the list
  embedding_vectors.append(embedding_vector)

# Print the list of embedding vectors
for vector in embedding_vectors:
  print(vector)
```

```
# prompt:  insert the chunk text and the corresponding vector to vector table

import sqlite3
import spacy

# Create a connection to the database
conn = sqlite3.connect('my_database.db')

# Create a cursor object
cursor = conn.cursor()

# Loop over each chunk of text and its corresponding vector
for chunk, vector in zip(chunks, embedding_vectors):
  # Insert the data into the vector_table
  cursor.execute("INSERT INTO vector_table (text, vector) VALUES (?, ?)", (chunk, vector))

# Save the changes to the database
conn.commit()

# Close the connection
conn.close()
```

```
# prompt: accept the user input and convert them in to vector in another table like user_vector table

# Create a connection to the database
conn = sqlite3.connect('my_database.db')

# Create a cursor object
cursor = conn.cursor()

# Create a table with two columns: name (text) and vector (array)
cursor.execute('''CREATE TABLE user_vector (name TEXT, vector ARRAY)''')

# Save the changes to the database
conn.commit()

# Close the connection
conn.close()

# Get the user input
user_input = input("Enter your text: ")

# Load the spaCy model
nlp = spacy.load('en_core_web_sm')

# Create a spaCy document object
doc = nlp(user_input)

# Get the embedding vector of the document
embedding_vector = doc.vector

# Create a connection to the database
conn = sqlite3.connect('my_database.db')

# Create a cursor object
cursor = conn.cursor()

# Insert the data into the user_vector table
cursor.execute("INSERT INTO user_vector (name, vector) VALUES (?, ?)", (user_input, embedding_vector))

# Save the changes to the database
conn.commit()

# Close the connection
conn.close()
```
